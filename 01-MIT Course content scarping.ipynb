```python
import csv
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get('https://ocw.mit.edu/courses/electrical-engineering-and-computer-science/6-006-introduction-to-algorithms-spring-2020/lecture-videos/#')
response.text
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
type(soup)
```

```python
all_links = soup.find_all("a", attrs={'class':"resource-list-title"})
len(all_links)
```

```python
name = all_links[0].text
url = "https://ocw.mit.edu" + all_links[0].attrs['href']
name, url
```

```python
with open('mit_alg_course.csv', 'w', newline='') as f:
    writer = csv.DictWriter(f, fieldnames=['Course Name', 'Link'])
    writer.writeheader()
    for link in all_links:
        name = link.text
        url = "https://ocw.mit.edu" + link.attrs['href']
        try:
            writer.writerow({'Course Name':name, 'Link':url})
        except:
            pass

import pandas as pd
df = pd.read_csv('mit_alg_course.csv')
df
```
