```python
import csv
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
response = requests.get('http://books.toscrape.com/catalogue/category/books/mystery_3/index.html')
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
all_articles = soup.find_all('article', attrs={'class': 'product_pod'})
len(all_articles)
```

```python

books = soup.find('ol', attrs={'class':'row'})
books = books.find_all('li')

with open('books.csv', 'w', newline='') as f:
    writer = csv.DictWriter(f, fieldnames=['title', 'price', 'rate'])
    writer.writeheader()
    for book in books:
        rates = {'One':1, 'Two':2, 'Three':3, 'Four':4, 'Five':5}
        stars_rate = book.find('p').attrs['class'][1]
        rate_values = rates[stars_rate]
        title = book.find_all('a')[1].get_text()
        price = float(book.find('p', attrs={'class':'price_color'}).get_text()[1:])

        writer.writerow({'title':title, 'price':price, 'rate':rate_values})

df = pd.read_csv('books.csv')
df
```

```python
def get_books(page):
    with open('books.csv', 'w') as f:
        writer = csv.DictWriter(f, fieldnames=['title', 'price', 'rate'])
        writer.writeheader()
        response = requests.request('GET', page)
        soup = BeautifulSoup(response.content, 'html.parser')
        all_articles = soup.find_all('article', attrs={'class': 'product_pod'})
        rates = {'One': 1, 'Two': 2, 'Three': 3, 'Four': 4, 'Five': 5}
        for article in all_articles:
            price = float(article.find('p', attrs={'class': 'price_color'}).get_text().split('£')[1])
            book_name = article.find('h3').find('a').get('title')
            rating = article.find('p').get('class')[1]
            writer.writerow({'title': book_name, 'price': price, 'rate': rates[rating]})
            print(f'Book name: {book_name}\nPrice: {price}\nRate: {rates[rating]}\n------')
```

```python
get_books('http://books.toscrape.com/catalogue/category/books/science_22/index.html')
```
